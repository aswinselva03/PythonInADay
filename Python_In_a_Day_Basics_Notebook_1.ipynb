## while loop

While loop will iterate through the code block until the condition is true



```
syntax:
while <condition>
    <statement1>
    <statement2>
    <statement3>
```

## while else loop


```
syntax:
while <condition>
    <statement1>
    <statement2>
    <statement3>
else                # else block will be executed when while condition fails
#    <statements>.   # else block will be skipped if while loop breaks with  break statement
```


image.png

````
# find even number from 0 to 10
i = 1

while i<=10:   #i<=10 is the condition
  if i%2 == 0:
    print(i)
  i+=1      # we need a increment or decrement statement inside the loop so that the
            # condition will failt at some point of time or else it will lead to an infinite loop

#iteration 1
# i = 1
# i%2 => 1%2 = 1 Does not print i

#iteration 2
# i = 2
# i%2 => 2%2 =0 Prints 2 as even
````

````
# break inside loop - print from 10 to 5
i = 10
while i >0:
  print(i)
  if i==5:
    break
  i -= 1
````

````
# continue inside loop - print only odd numbers
i = 1
while i < 10:
  if i%2 == 0:
    i+=1
    continue
  print(i)
  i+=1
````

````
# while else
i = 1
# num = 11
num = 21
n = 20
while i<=n:
  if i%num == 0:
    print(str(i)+ " is the first number divisible by " + str(num))
    break
  i+=1
else:
  print("No number is divisible by "+str(num)+ " in the range from 1 to " + str(n))


#while else with break

````

## Iterators in python

An iterator is an object that contains a countable number of values. An iterator is an object that can be iterated upon, meaning that you can traverse through all the values.


```
example:

range(1,11) - this creates an iterator with values from 1 to 10, which can be looped using for loop

```

````
# print even numbers from 1 to 10

for i in range(1,11):
  if i%2 == 0:
    print(i)

#iteration 1
# i = 1
# i%2 => 1%2 = 1 Does not print i

#iteration 2
# i = 2
# i%2 => 2%2 =0 Prints 2 as even
````

````
# print numbers from 1 to 6 using for loop and break
for i in range(1,11):
  print(i)
  if i == 6:
    break
````

````
# print odd numbers using for loop and continue
for i in range(1,11):
  if i%2 == 0:
    continue
  print(i)
````

````
# for else loop
name = "Brynn"
# name = "Casey"

for i in name:
  if i in "aeiou":
    print(name+" has vowel in it")
    break
else:
  print(name+" does not have vowel in it")
````
